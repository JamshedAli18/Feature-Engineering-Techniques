```python
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import numpy as np

```

```python
# Load dataset
car_data = fetch_openml(name="car", version=1, as_frame=True)
df = car_data.frame

```

```python
df.head()
```

```python
# Ensure y is a pandas Series
y = df['class']

# Select categorical columns
categorical_cols = df.select_dtypes(include=["object"]).columns

```

```python
# 1. Label Encoding
le = LabelEncoder()
df['buying_Label'] = le.fit_transform(df['buying'])

```

```python
df.head()
```

```python
# 2. One-Hot Encoding
df_one_hot = pd.get_dummies(df, columns=["maint"], drop_first=True)

```

```python
# 3. Frequency Encoding
freq_encoding = y.value_counts().to_dict()
df['class_Freq_Encoded'] = y.map(freq_encoding)

```

```python
df
```

```python
# 4. Target Encoding (Manual Approach)
# Calculate target mean for each category
target_mean = df.groupby('buying')['class'].apply(lambda x: x.mode()[0] if not x.mode().empty else np.nan).to_dict()
df['buying_Target_Encoded'] = df['buying'].map(target_mean)
```
